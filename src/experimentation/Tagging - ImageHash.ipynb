```python
from sprite_tester import SpriteTester

from PIL import Image
import imagehash
from scipy.misc import imresize
import cv2
```

```python
def hasher(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = cv2.resize(img, (20, 20))
    _, img = cv2.threshold(img,16,255,0)
    
    return imagehash.average_hash(Image.fromarray(img))

def imageHashDifference(hash1, hash2):
    return hash1 - hash2

t = SpriteTester(hasher, imageHashDifference)
(p, sub, r) = t.runTests()
(p, sub, [a for a in r if a[1] == 'electrode'])
```

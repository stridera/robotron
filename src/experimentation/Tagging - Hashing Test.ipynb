```python
from sprite_tester import SpriteTester
import imagehash
import cv2
from PIL import Image
from scipy.misc import imresize
                
def hasher(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = cv2.resize(img, (20, 20))
#     img = cv2.GaussianBlur(img, (3, 3), 0)
    _, img = cv2.threshold(img,16,255,0)
    
    return imagehash.average_hash(Image.fromarray(img))

def imageHashDifference(hash1, hash2):
    return hash1 - hash2

t = SpriteTester(hasher, imageHashDifference)
(p, sub, r) = t.runTests()
print(p)
# (p, sub, [a for a in r if a[0] == 'electrode'])
```

```python
names, hashes = t.getHashes()

hashMap = []
for i in range(15):
    hashMap.append([])
    
for n, h in zip(names, hashes):
    h = str(h)
    c = t.getClass(n)
    print(c, n)
    if not h in hashMap[int(c)]:
        hashMap[int(c)].append(h)

print(hashMap)
for i in range(15):
    print(i, hashMap[i])
```
